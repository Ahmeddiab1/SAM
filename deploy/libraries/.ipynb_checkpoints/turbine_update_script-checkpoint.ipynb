```python
sheet_names = pd.DataFrame(new_data.sheetnames)
#print(sheet_names.iloc[0,0])
check_array= pd.DataFrame([],index=new_wave_library.index)
check_array['Name Check']=np.nan
check_array['New Wind Speed Array']=np.nan
check_array['New Power Curve Array']=np.nan
worksheet_loop = np.linspace(1,sheet_names.index[-6],11,endpoint=True)
#print(worksheet_loop)
for j in worksheet_loop:
    b = int(j)
    #print(b)
    sheet_data = sheet_names.iloc[b,0]

    #print(sheet_data)
    ws = new_data[str(sheet_data)]
    ws_df = pd.DataFrame(ws.values)
    ws_df = ws_df.round(3)
    if isinstance(ws_df.iloc[5,0],str)==True:
        start_var = 6
    else:
        start_var = 5
    #print(ws_df)
    loop_array = np.linspace(start_var,ws_df.index[-1]+1,ws_df.index[-1]-start_var+1,endpoint=False)
    print(loop_array)
    for i in loop_array:
            a = int(i)
            #print(a)
            if ws_df.iloc[a,0] is None:
                ws_df.iloc[a,0]=9999
            if ws_df.iloc[a,5] is None:
                ws_df.iloc[a,5]=9999
            if ws_df.iloc[a,1] is None:
                ws_df.iloc[a,1]=9999    
            if ws_df.iloc[a,6] is None:
                ws_df.iloc[a,6]=9999
            if i==start_var:
                old_wind_speed_string = str('%5.3d' % int(ws_df.iloc[a,0]))+"|"
                new_wind_speed_string = str('%5.3d' % ws_df.iloc[a,5])+"|"
                old_power_curve_string = str('%5.3d' % ws_df.iloc[a,1])+"|"
                new_power_curve_string = str('%5.3d' % ws_df.iloc[a,6])+"|"
            elif i==ws_df.index[-1]:
                old_wind_speed_string = old_wind_speed_string + str('%5.3d' % ws_df.iloc[a,0])
                new_wind_speed_string = new_wind_speed_string + str('%5.3d' % ws_df.iloc[a,5])
                old_power_curve_string = old_power_curve_string + str('%5.3d' % ws_df.iloc[a,1])
                new_power_curve_string = new_power_curve_string + str('%5.3d' % ws_df.iloc[a,6])
            else:
                old_wind_speed_string = old_wind_speed_string + str('%5.3d' % ws_df.iloc[a,0])+"|"
                new_wind_speed_string = new_wind_speed_string + str('%5.3d' % ws_df.iloc[a,5])+"|"
                old_power_curve_string = old_power_curve_string + str('%5.3d' % ws_df.iloc[a,1])+"|"
                new_power_curve_string = new_power_curve_string + str('%5.3d' % ws_df.iloc[a,6])+"|"
            old_wind_speed_string = old_wind_speed_string.replace('9999|',"")
            new_wind_speed_string = new_wind_speed_string.replace('9999|',"")
            old_power_curve_string = old_power_curve_string.replace('9999|',"")
            new_power_curve_string = new_power_curve_string.replace('9999|',"")
            old_wind_speed_string = old_wind_speed_string.replace('|9999',"")
            new_wind_speed_string = new_wind_speed_string.replace('|9999',"")
            old_power_curve_string = old_power_curve_string.replace('|9999',"")
            new_power_curve_string = new_power_curve_string.replace('|9999',"")
    check_array_values = new_wave_library[new_wave_library['Power Curve Array']==old_power_curve_string]# & new_wave_library['Power_Curve_Array'].str.contains[old_power_curve_string]]
    check_array.iloc[check_array_values.index,0]=str(sheet_data)
    check_array.iloc[check_array_values.index,1]=new_wind_speed_string
    check_array.iloc[check_array_values.index,2]=new_power_curve_string
print(old_wind_speed_string)
print(new_wind_speed_string)
print(new_power_curve_string)
print(new_wave_library['Power Curve Array'])
print(old_power_curve_string)
print(check_array)

print(new_wave_library)
#print(new_wind_speed_string)
#print(ws_df)
#for i in sheet_names.index:
 #   sheet_data = sheet_names.iloc[i,0]
  #  ws = new_data[str(sheet_data)]
   # ws_df = pd.DataFrame(ws.values)
    
```

```python
#New Turbine Power arrays input in different format than modified array values
worksheet_loop2 = np.linspace(sheet_names.index[-5],sheet_names.index[-1],5,endpoint=True)
print(worksheet_loop2)
for k in worksheet_loop2:
    c = int(k)
    print(c)
    sheet_data = sheet_names.iloc[c,0]

    #print(sheet_data)
    ws = new_data[str(sheet_data)]
    ws_df = pd.DataFrame(ws.values)
    ws_df = ws_df.round(3)
    #print(ws_df)
    if isinstance(ws_df.iloc[5,0],str)==True:
        start_var=6
    else:
        start_var=5
    loop_array2 = np.linspace(start_var,ws_df.index[-1]+1,ws_df.index[-1]-start_var+1,endpoint=False)
    for m in loop_array2:
            d = int(m)
            #print(d)
            column2_bool = (ws_df.iloc[:,2].str.contains('Hub Height'))
            #column3_bool = (ws_df['3'].str.contains('Hub Height'))
            if any(column2_bool==True):
                if ws_df.iloc[d,2] is None:
                    ws_df.iloc[d,2]=9999
                if ws_df.iloc[d,3] is None:
                    ws_df.iloc[d,3]=9999
                if m==start_var:
                    new_wind_speed_string = str('%5.3f' % ws_df.iloc[d,2])+"|"
                    new_power_curve_string = str('%5.3f' % ws_df.iloc[d,3])+"|"
                elif m==ws_df.index[-1]:
                    new_wind_speed_string = new_wind_speed_string + str('%5.3f' % ws_df.iloc[d,2])
                    new_power_curve_string = new_power_curve_string + str('%5.3f' % ws_df.iloc[d,3])
                else:
                    new_wind_speed_string = new_wind_speed_string + str('%5.3f' % ws_df.iloc[d,2])+"|"
                    new_power_curve_string = new_power_curve_string + str('%5.3f' % ws_df.iloc[d,3])+"|"
            else:
                if ws_df.iloc[d,1] is None:
                    ws_df.iloc[d,1]=9999
                if ws_df.iloc[d,2] is None:
                    ws_df.iloc[d,2]=9999
                if m==start_var:
                    new_wind_speed_string = str('%5.3f' % ws_df.iloc[d,1])+"|"
                    new_power_curve_string = str('%5.3f' % ws_df.iloc[d,2])+"|"
                elif m==ws_df.index[-1]:
                    new_wind_speed_string = new_wind_speed_string + str('%5.3f' % ws_df.iloc[d,1])
                    new_power_curve_string = new_power_curve_string + str('%5.3f' % ws_df.iloc[d,2])
                else:
                    new_wind_speed_string = new_wind_speed_string + str('%5.3f' % ws_df.iloc[d,1])+"|"
                    new_power_curve_string = new_power_curve_string + str('%5.3f' % ws_df.iloc[d,2])+"|"
            new_wind_speed_string = new_wind_speed_string.replace('9999|',"")
            new_power_curve_string = new_power_curve_string.replace('9999|',"")
            new_wind_speed_string = new_wind_speed_string.replace('|9999',"")
            new_power_curve_string = new_power_curve_string.replace('|9999',"")
    check_array_values = new_wave_library[new_wave_library['Name'].str.contains(sheet_data)]# & new_wave_library['Power_Curve_Array'].str.contains[old_power_curve_string]]
    check_array.iloc[check_array_values.index,0]=str(sheet_data)
    check_array.iloc[check_array_values.index,1]=new_wind_speed_string
    check_array.iloc[check_array_values.index,2]=new_power_curve_string
new_wave_library['Name Check']=check_array['Name Check']
new_wave_library['New Wind Speed Array']=check_array['New Wind Speed Array']
new_wave_library['New Power Curve Array']=check_array['New Power Curve Array']
new_wave_library.to_csv('C:/Users/mprillim/sam_dev/sam/deploy/libraries/wind_library_test2.csv')
```
